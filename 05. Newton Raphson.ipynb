```
def truncate(floatt, n):
  '''truncate floatt to n digits'''
  return int(floatt * 10**n)/10.0**n

def checkfor(fx, x, xvalans, maxmin):
  d2fx = diff(fx, x, 2).subs(x, xvalans)
  print(d2fx, end='\t')

  if maxmin=='maximum':
    pass
  elif maxmin=='mimimum':
    d2fx = d2fx*-1

  print(d2fx<0)
  

def newtonraphson(fx, x, n, xn, stoppingcrit, limititer, typee):
  dfx = diff(fx, x, 1) # first derivative
  print('fx  =',fx)
  print('dfx =',dfx)

  print('\nI\txn\txn+1\tError')
  for i in range(1,limititer+2):
    xn1 = truncate(xn - fx.subs(x,xn)/dfx.subs(x,xn), n)

    if typee==1:
      error = (stoppingcrit)
    if typee==2:      
      error = (round(abs(xn1-xn), n), stoppingcrit)
    if typee==3:      
      error = (truncate(abs((xn1-xn)/xn1), n), stoppingcrit)

    print(f'{i}\t{xn}\t{xn1}\t{error}')
    xn = xn1
```

```
x = Symbol('x')

fx = e**(-x) - x
stoppingcrit = 0.5/100
n = 4
initialguess = 0
typee = 3
limititer = 5

newtonraphson(fx, x, n, initialguess, stoppingcrit, limititer, typee)
```
