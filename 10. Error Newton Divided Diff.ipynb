```
from math import ceil, pi, e, sqrt
from sympy import series, Symbol, Rational, diff, factorial, fraction, nsimplify, solve, S, ln, expand
from sympy.functions import sin, cos, tan
import pandas as pd
from IPython.display import display, HTML
```

```
def truncate(floatt):
  '''truncate floatt to n digits'''
  return int(floatt * 10000)/10000.0

def errornewtondivideddiff(coords, xs, additionalpoint):
  x = Symbol('x')

  values = [[i[1] for i in coords]]
  length = len(coords)
  difference = 1
  for i in range(len(coords)-1):
    temp = list()
    orders = values[-1]
    for j in range(len(orders)-1):
      temp.append(truncate((orders[j+1]-orders[j])/(xs[j+difference]-xs[j])))
    difference+=1
    values.append(temp)

  # creating table
  data = {'x':xs,'y':values[0]}
  [data.update({i:values[i]}) for i in range(1,length)]

  df = pd.DataFrame.from_dict(data, orient='index')
  df = df.transpose().fillna('-') # table created
  display(HTML(df.to_html(index=False)))

  xval = list()
  gx = 1
  for i in xs:
    if i!=additionalpoint:
      xval.append(x-i)
      gx*=xval[-1]
  
  gx = expand(gx)
  print(f'\n|R{length-2}x| = |{values[-1][0]}/({length-2}+{1})!) * ',end='')
  [print(f'({i})',end='') for i in xval]
  print(f'|\n|R{length-2}x| = |{values[-1][0]}/{factorial(length-1)}| * |{gx}|\n')

  print(f'gx = {gx}') # finding maximum value of gx
  print(f'g\'x = {diff(gx,x,1)} = 0')
  vals  = solve(diff(gx,x,1),x) # x values found here
  vals = [nsimplify(i) for i in vals] # put them in fractions
  print(f'x = {vals}')
  vals2 = [truncate(abs(gx.subs(x,i))) for i in vals] # truncate to 4 decimals

  [print(f'|g({vals[i]})| = {vals2[i]}') for i in range(len(vals))] # equation printing
  maxx = max(vals2)
  print('\n(max) =',maxx, end='\n')

  cpart = values[-1][0]/factorial(length-1)

  print(f'\nR{length-2}x <= {values[-1][0]}/{factorial(length-1)} * {maxx}')
  print(f'R{length-2}x <= {truncate(cpart*maxx)}')
```

```
coords = [(1,0), (4,1.3862), (5,1.6093), (6,1.7917)]
additionalpoint = 5
xs = [i[0] for i in coords]

errornewtondivideddiff(coords, xs, additionalpoint)
```
