```
from math import ceil, pi, e
from sympy import series, Symbol, Rational, diff, factorial, fraction, nsimplify, solve, S, ln, expand
from sympy.functions import sin, cos, tan
```

```
def truncate(floatt):
  '''truncate floatt to n digits'''
  return int(floatt * 10000)/10000.0

def errorlagrange(fx, xs, a, b, n):
  c = Symbol('c')
  listt = [(x-i) for i in xs]

  print(f'x = {xs}\nfx = {fx}\na,b = [{a},{b}]\nn = {n}\nR{n}x = ?\n') # given

  print(f'R{n}x = f({n}+1)(c)/({n}+1)! * ',end='') # equation printing
  [print(f'(x-x{i})',end='') for i in range(len(listt))]

  print(f'\nR{n}x = f({n+1})(c)/({n+1})! * ',end='') # equation printing
  [print(f'({i})',end='') for i in listt]
  print('\n')

  cpart = diff(fx,x,n+1).subs(x,c)/factorial(n+1)

  for i in range(n+1):
    print(f'{i+1} derivative = {diff(fx,x,i+1)}') # finding derivatives

  temp = 1
  for i in listt:
    temp *=i # multiplying x terms

  print(f'\nR{n}x = |{cpart}| * |',end='') # equation printing
  [print(f'({i})',end='') for i in listt]

  gx = expand(temp)
  print(f'|\nR{n}x = |{cpart}| * |{gx}|',end='') # equation printing

  temp *= cpart

  print(f'\n\ngx = {gx}') # finding maximum value of gx
  print(f'g\'x = {diff(gx,x,1)} = 0')
  vals  = solve(diff(gx,x,1),x) # x values found here
  vals = [nsimplify(i) for i in vals] # put them in fractions
  print(f'x = {vals}')
  vals2 = [truncate(abs(gx.subs(x,i))) for i in vals] # truncate to 4 decimals

  [print(f'|g({vals[i]})| = {vals2[i]}') for i in range(len(vals))] # equation printing
  maxx = max(vals2)
  print('\n(max) =',maxx, end='\n')
  
  aval = abs(cpart.subs(c,a)) # max value between a, b
  bval = abs(cpart.subs(c,b))
  if aval>bval:
    print(f'\nmax value is on c = {a}, {aval}\n')
    print(f'R{n}x <= ({aval}) * {maxx} = {truncate(aval*maxx)}')
  else:    
    print(f'\nmax value is on c = {b}, {bval}\n')
    print(f'R{n}x <= ({bval}) * {maxx} = {truncate(bval*maxx)}')
  
```

EXAMPLE

```
x = Symbol('x')

xs = [2,2.75,4]
fx = 1/x
a,b = 2,4       #interval
n = len(xs)-1

errorlagrange(fx, xs, a, b, n)
```
