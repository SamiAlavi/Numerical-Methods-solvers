```
from math import ceil, pi, e
from sympy import Symbol, symbols, solve
from termcolor import colored
import numpy as np
```

```
def truncate(floatt,n):
  floatt = str(float(floatt))
  indx = floatt.find('.')
  if indx!=-1:
    try:
      return float(floatt[:indx+n+1])
    except:
      return float(floatt)
  else:
    return float(floatt)

def printt(string,color="red",end='\n'):
  print(colored(string,color),end=end)

def finderror(xs, k, n):
  arr1, arr2 = np.array(xs[k-1]), np.array(xs[k])
  subt = np.absolute(arr1 - arr2)
  num = subt.max()
  indx = np.where(subt==num)
  den = arr2.max()
  error = truncate(num/den,n)
  print(f'Error = ({arr2[indx][0]} - {arr1[indx][0]}) / {den} = ',end='')
  printt(f'{truncate(num,n)}/{den} = {error}',color='blue')
  printt('----------------------------------------------')
  return error
```

```
def jaccobimethod(eqs, symbs, xs, n, errorlim, numiters):
  flag = errorlim!=None
  xs = [xs]
  length = len(eqs)
  values = dict()
  errorlist = ['-']
  [values.update({symbs[i]:xs[0][i]}) for i in range(length)]

  for i in range(length):
    eqs[i] = solve(eqs[i],symbs[i])[0]
    print(f'{symbs[i]} = {eqs[i]}')
  print()

  for k in range(1,numiters+1):
    printt(f'k = {k}')
    xlist = list()
    for i in range(length):
      ans = eqs[i].subs(values)
      fltans = truncate(ans, n)
      xlist.append(fltans)
      print(f"{symbs[i]} = {ans} = {colored(f'{fltans}','blue')}")
    xs.append(xlist)
    [values.update({symbs[i]:xs[k][i]}) for i in range(length)]
    errorlist.append(finderror(xs,k,n))
    if flag:
      if errorlist[-1]<errorlim:
        break

  stringg = '\t'.join([str(i) for i in symbs])
  printt(f'k\t{stringg}\tError')
  for i in range(len(xs)):
    if i%2==0:
      color = 'blue'
    else:
      color = 'green'
    printt(i,color=color,end='\t')
    for j in range(length):
      printt(xs[i][j],color=color,end='\t')
    printt(errorlist[i],color=color)
  print(f'\n{symbs} = {xs[-1]}')
```

```
x1, x2, x3, x4 = symbols('x1 x2 x3 x4') # variables
symbs = [x1,x2,x3,x4]                   # equations
eqs = [10*x1 - x2 + 2*x3 - 6,
       -x1 + 11*x2 - x3 + 3*x4 - 25,
       2*x1 - x2 + 10*x3 - x4 + 11,
       3*x2 - x3 + 8*x4 - 15]
xs = [0,0,0,0]                          # initial x values
errorlim = 0.001                        # if error lim, else set to None
n = 4                                   # truncate upto n decimals
numiters = 50                           # num of iterations (set to 50 if to find using errorlim)

jaccobimethod(eqs, symbs, xs, n, errorlim, numiters)
```
