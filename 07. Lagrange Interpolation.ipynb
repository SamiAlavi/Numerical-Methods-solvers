```
from math import ceil, pi, e
from sympy import series, Symbol, Rational, diff, factorial, fraction, nsimplify, solveset, S, ln, expand, fraction, together, N
from sympy.functions import sin, cos, tan
from termcolor import colored
```

```
def printt(string, color, end='\n'):
  print(colored(string,color),end=end)

def fitlagrange(coords, x, xval):
  printt('Red can be ignored as its y value is 0','red')
  Lx = list()
  strilist = list()
  for i in range(len(coords)):
    tempnum = 1
    tempden = 1
    string = ''
    for j in range(len(coords)):
      if j!=i:
        tempnum = tempnum * (x - coords[j][0])
        tempden = tempden * (coords[i][0] - coords[j][0])
        string += f'({coords[i][0]} - {coords[j][0]})'
    Lx.append(tempnum/tempden)
    if coords[i][1]!=0:
      strilist.append(f'({tempnum}) / ({string})')
    else:
      strilist.append(colored(f'({tempnum}) / ({string})','red'))
  printt('Lx','green')
  [print(f"Lx{i} = {strilist[i]}\t{colored(f'=> {Lx[i]}','blue')}") for i in range(len(Lx))]

  printt('\nLx (expanded)','green')
  if xval==None:
    for i in range(len(Lx)):
      temp = fraction(together(Lx[i]))
      if coords[i][1]!=0:
        print(f'Lx{i} = ({expand(temp[0])}) / {temp[1]}')
      else:
        printt(f'Lx{i} = ({expand(temp[0])}) / {temp[1]}','red')
  else:
    [print(f'Lx{i} = ',Lx[i].subs(x,xval)) for i in range(len(Lx))]


  Px = list()
  for i in range(len(coords)):
    Px.append(Lx[i] * coords[i][1])
  printt(f'\nPx (expanded)\nPx{len(Px)} = ','green',end='')
  [print(f'Lx{i}*y{i} + ',end='') for i in range(len(Px))]
  print('\nPx =',end='\t')

  if xval == None:
    [print(f'({Lx[i]})({coords[i][1]}) +',end=' ') for i in range(len(Px))]
    print()
    for i in range(len(Px)):
      temp = fraction(together(Px[i]))
      print(f'Px{i} = ({expand(temp[0])}) / {temp[1]}')
  else:
    [print(f'({Lx[i].subs(x,xval)})({coords[i][1]}) +',end=' ') for i in range(len(Px))]
    print()
    [print(f'Px{i} = ',Px[i].subs(x,xval)) for i in range(len(Px))]

  ans = sum(Px)
  if xval == None:
    printt(f'\n\nPx = \t{ans}','red')
    print(f'Px = \t{expand(ans)}')
  else:
    ans = ans.subs(x,xval)
    printt(f'\n\nPx = \t{ans} => {float(ans)}','red')
```

```
x = Symbol('x')

coords = [(1,-3), (3,0), (4,30), (6,132)]
xval = None

fitlagrange(coords, x, xval)
```
