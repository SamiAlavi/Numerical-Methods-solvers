```
from math import ceil, pi, e
from sympy import series, Symbol, Rational, diff, factorial, fraction, nsimplify, solveset, S, ln, expand
from sympy.functions import sin, cos, tan
import pandas as pd
from IPython.display import display, HTML
```

```
def truncate(floatt):
  '''truncate floatt to n digits'''
  return int(round(floatt,10) * 10000)/10000.0

def newtondivideddiff(coords, xval):
  xs = [i[0] for i in coords]
  values = [[i[1] for i in coords]]
  length = len(coords)
  diff = 1
  for i in range(len(coords)-1):
    temp = list()
    orders = values[-1]
    for j in range(len(orders)-1):
      temp.append(truncate((orders[j+1]-orders[j])/(xs[j+diff]-xs[j])))
    diff+=1
    values.append(temp)

  # creating table
  data = {'x':xs,'y':values[0]}
  [data.update({i:values[i]}) for i in range(1,length)]

  df = pd.DataFrame.from_dict(data, orient='index')
  df = df.transpose().fillna('-') # table created
  for i in range(1,length):
    df[i] = df[i].astype(str)
  display(HTML(df.to_html(index=False)))

  # solving Px
  x = Symbol('x')
  Px = [values[0][0]]

  for i in range(1,length):
    temp = 1
    print(f'\nP{i}x = ({Px[-1]}) +',end=' ')
    for j in range(i):
      temp *= (x-xs[j])
      print(f'({x-xs[j]})', end='')
    temp *= values[i][0]
    print(f'({values[i][0]})\nP{i}x = {Px[-1]} + {temp}')
    Px.append(expand(Px[-1] + temp))
    print(f'P{i}x =',Px[-1])
    if xval!=None:
      print(f'P{i}x =',truncate(Px[-1].subs(x,xval)))
  
```

EXAMPLE

```
coords = [(-0.1,5.3), (0,2), (0.2,3.19), (0.3,1)]
xval = None

newtondivideddiff(coords, xval)
```
