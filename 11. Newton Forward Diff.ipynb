```
from math import ceil, pi, e
from sympy import series, Symbol, Rational, diff, factorial, fraction, simplify, solveset, S, ln, expand, N
from sympy.functions import sin, cos, tan
import pandas as pd
from IPython.display import display, HTML
```

```
def truncate(floatt):
  '''truncate floatt to n digits'''
  return round(round(floatt,10)*100000)/100000.0

def newtonforwdivideddiff(coords, xs, x, n):
  values = [[i[1] for i in coords]]
  length = len(coords)
  h = truncate(xs[1] - xs[0])
  flag = x==None
  if flag:
    x = Symbol('x')
    s = (x - xs[0])/h
  else:
    s = truncate((x - xs[0])/h)
  print(f's = (x-xs[0])/(h)')
  print(f's = ({x}-{xs[0]})/({xs[1]}-{xs[0]})\ns = {s}')

  for i in range(len(coords)-1):
    temp = list()
    orders = values[-1]
    for j in range(len(orders)-1):
      temp.append(truncate(orders[j+1]-orders[j]))
    values.append(temp)

  # creating table
  data = {'x':xs,'y':values[0]}
  [data.update({i:values[i]}) for i in range(1,length)]

  df = pd.DataFrame.from_dict(data, orient='index')
  df = df.transpose().fillna('-') # table created
  display(HTML(df.to_html(index=False)))

  if flag:  # for equation finding
    findequation(values, s, length)
  else:  # for solution finding
    findsolution(values, s, length)
    

def findequation(values, s, length):
  Px = [values[0][0], s]
  sol = list()

  print(f'\ny0 = ({Px[0]})')
  print(f'\n(s-0) =\t{s}', end='')
  for i in range(1,n):
    temp = Px[-1]
    print('\n')
    for j in range(i):
      print(f'(s-{i-j})', end='')
    temp *= (s-i)
    Px.append(expand(temp))
    print(f' = {temp} \n = {Px[-1]}',end='')

  print(f'\n\nP{n}x =\t({Px[0]}) +\n', end='')
  sol.append(Px[0])
  for i in range(1,len(Px)):
    print(f'\t({Px[i]})({values[i][0]})/{i}! +\n ',end='')
    temp = Px[i]*values[i][0]/factorial(i)
    sol.append(N(temp, 5 + ln(i, 10)))
  
  print(f'\nP{n}x =', end='')
  [print(f'\t({i}) +\n ',end='') for i in sol]
  print(f'\nP{n}x = {sum(sol)}')

def findsolution(values, s, length):
  Px = [values[0][0]]

  print(f'\nP{length-1}x =\t({Px[-1]})', end='')
  for i in range(1,length):
    temp = 1
    print(f' + \n\t',end='')
    for j in range(i):
      temp *= (s-j)
      print(f'({s}-{j})', end='')
    temp *= values[i][0]/factorial(j+1)
    print(f'({values[i][0]})/{j+1}!',end='')
    Px.append(truncate(temp))
  
  print(f'\n\nP{length-1}x =\t', end='')
  [print('{:.5f} +'.format(i),end='\n\t') for i in Px]

  print(f'\nP{length-1}x = {sum(Px)}')
```

EXAMPLE

```
coords = [(0.1,0.003),(0.3,0.067),(0.5,0.148),(0.7,0.248),(0.9,0.370),(1.1,0.518),(1.3,0.698)]
x = 0.213
n = None
xs = [i[0] for i in coords]

newtonforwdivideddiff(coords, xs, x, n)
```
