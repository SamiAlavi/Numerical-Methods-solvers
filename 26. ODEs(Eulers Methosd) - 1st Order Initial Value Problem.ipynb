```
from math import ceil, pi, e
from sympy import Symbol, diff, factorial, fraction, simplify, solve, N, symbols
from sympy.functions import sin, cos, tan
from termcolor import colored
```

```
def checkint(floatt):
  floatt = float(floatt)
  if floatt%1!=0:
    return floatt
  else:
    return int(floatt)

def truncate(floatt,n):
  floatt = str(float(floatt))
  indx = floatt.find('.')
  if indx!=-1:
    try:
      return checkint(floatt[:indx+n+1])
    except:
      return checkint(floatt)
  else:
    return checkint(floatt)

def printt(string,color="red",end='\n'):
  print(colored(string,color),end=end)
```

```
def eulermethod(fx,h,xvals,interval,n):
  a,b = interval
  N = round((b-a)/h)
  ys =  [xvals[1]]
  ts = [a]
  dictt = {'y':f'yi', 't':f'ti'}
  for i in range(N):ts.append(truncate(ts[-1]+h,n))
  print(f'h = (b-a)/N =>   N = ({b}-{a})/{h} => {N}')
  [print(f't{i}={ts[i]}, ',end='') for i in range(len(ts))]
  print(f'\nf(t,y) = {fx}')
  printt(f'\nEuler method is\ny({xvals[0]}) = y(t0) = y0 = {ys[0]}\nyi+1 = y(ti+1) = yi + (h)f(ti,yi)\nyi+1 = y(ti+1) = yi + ({h})({fx.subs(dictt)})\n')

  for i in range(1,N+1):
    printt(f'i={i})')
    dictt = {'y':f'y{i-1}', 't':f't{i-1}'}
    dictvals = {'y':ys[i-1], 't':ts[i-1]}
    val = truncate(fx.subs(dictvals),n)
    ys.append(truncate(ys[i-1] + h*val,n))
    print(f'y{i} = y(t{i}) = y{i-1} + ({h})({fx.subs(dictt)}) ',end='')
    printt(f"=> {ys[i-1]} + ({h})({str(fx).replace('y',str(ys[i-1])).replace('t',str(ts[i-1]))}) ",color='blue',end='')
    print(f"=> {ys[i-1]} + ({h})({val}) ",end='')
    printt(f'= {ys[-1]}',color='green')

  printt('\nSolution')
  print('t',end='\t')
  [print(i,end='\t') for i in ts]
  print('\ny',end='\t')
  [print(i,end='\t') for i in ys]
```

```
y,t = symbols('y t')    # set symbols

fx = y - t**2 + 1       # equation
h = 0.5
xvals = (0,0.5)         # y(0) = 0.5
interval = (0,2)        # set interval
n = 4                   # set decimals to truncate

eulermethod(fx,h,xvals,interval,n)
```

```
y,t = symbols('y t')    # set symbols

fx = e**(t-y)           # equation
h = 0.25
xvals = (0,0.5)         # y(0) = 0.5
interval = (0,1)        # set interval
n = 4                   # set decimals to truncate

eulermethod(fx,h,xvals,interval,n)
```
