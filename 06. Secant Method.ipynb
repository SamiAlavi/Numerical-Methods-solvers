```
def truncate(floatt, n):
  '''truncate floatt to n digits'''
  return int(floatt * 10**n)/10.0**n

def checkfor(fx, x, xvalans, maxmin):
  d2fx = diff(fx, x, 2).subs(x, xvalans)
  print(d2fx, end='\t')

  if maxmin=='maximum':
    pass
  elif maxmin=='mimimum':
    d2fx = d2fx*-1

  print(d2fx<0)
  

def secantmethod(fx, x, n, xns, stoppingcrit, limititer, typee):

  print('\nI\txn\txn+1\tf(xn)\tf(xn+1)\txn+2\tError')
  for i in range(1,limititer+1):
    fxn = truncate(fx.subs(x,xns[0]), n)
    fxn1 = truncate(fx.subs(x,xns[1]), n)
    xns[2] = (xns[0]*fxn1 - xns[1]*fxn)/(fxn1 - fxn)
    xns[2] = truncate(xns[2], n)

    if typee==1:
      error = (stoppingcrit)
    if typee==2:      
      error = (round(abs(xns[2]-xns[1]), n), stoppingcrit)
    if typee==3:      
      error = (truncate(abs((xns[2]-xns[1])/xns[2]), n), stoppingcrit)

    print(f'{i}\t{xns[0]}\t{xns[1]}\t{fxn}\t{fxn1}\t{xns[2]}\t{error}')
    
    xns[0], xns[1] = xns[1], xns[2]
```

```
x = Symbol('x')

fx = e**(-x) - x
stoppingcrit = 0.1/100
n = 4
initialguess = [-1, 0, None]
typee = 3
limititer = 5

secantmethod(fx, x, n, initialguess, stoppingcrit, limititer, typee)
```
