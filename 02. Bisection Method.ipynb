Q) Show that 𝑓(𝑥)=𝑥^3+4𝑥^2−10=0 has root in the interval [1,2]. Use bisection method to determine approximation to the root that is accurate to at least within 10^(−3).

```
x = Symbol('x')

fx = x**3 + 4*x**2 - 10 
(a, b) = (1 , 2)
n = 3

stoppingcrit = 10**(-n)
limititer = 15
typee = 1

brackmethod(fx, a, b, typee, stoppingcrit, limititer)
```
