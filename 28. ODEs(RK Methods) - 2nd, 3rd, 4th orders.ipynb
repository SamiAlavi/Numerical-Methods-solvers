```
from math import ceil, pi, e
from sympy import Symbol, diff, factorial, fraction, simplify, solve, N, symbols
from sympy.functions import sin, cos, tan
from termcolor import colored
```

```
def checkint(floatt):
  floatt = float(floatt)
  if floatt%1!=0:
    return floatt
  else:
    return int(floatt)

def truncate(floatt,n):
  floatt = str(float(floatt))
  indx = floatt.find('.')
  if indx!=-1:
    try:
      return checkint(floatt[:indx+n+1])
    except:
      return checkint(floatt)
  else:
    return checkint(floatt)

def printt(string,color="red",end='\n'):
  print(colored(string,color),end=end)
```

```
def getks(fx,h,t,y,n,order=2):
  ks = list()
  temp = str(fx)
  values = {'t':t,'y':y}
  k1 = truncate(fx.subs(values),n)
  print(f"k1 = {temp.replace('t',f'({t})').replace('y',f'({y})')} = {k1}")
  ks.append(k1)
  if order>2:
    values = {'t':t+h/2,'y':y+(h/2)*k1}
    k2 = truncate(fx.subs(values),n)
    print(f"k2 = {temp.replace('t',f'({t}+{h}/2)').replace('y',f'({y}+({h}/2)*{k1})')}")
    printt(f"   = {temp.replace('t',f'({truncate(t+h/2,n)})').replace('y',f'({truncate(y+(h/2)*k1,n)})')} = {k2}",color='blue')
    ks.append(k2)
  if order==3:
    values = {'t':t+h,'y':y-h*k1+2*h*k2}
    k3 = truncate(fx.subs(values),n)
    print(f"k3 = {temp.replace('t',f'({t}+{h})').replace('y',f'({y}-{h}*{k1}+2*{h}*{k2})')}")
    printt(f"   = {temp.replace('t',f'({truncate(t+h,n)})').replace('y',f'({truncate(y-h*k1+2*h*k2,n)})')} = {k3}",color='blue')
    ks.append(k3)
  elif order==4:
    values = {'t':t+h/2,'y':y+(h/2)*k2}
    k3 = truncate(fx.subs(values),n)
    print(f"k3 = {temp.replace('t',f'({t}+{h}/2)').replace('y',f'({y}+({h}/2)*{k2})')}")
    printt(f"   = {temp.replace('t',f'({truncate(t+h/2,n)})').replace('y',f'({truncate(y+(h/2)*k2,n)})')} = {k3}",color='blue')
    ks.append(k3)
    
    values = {'t':t+h,'y':y+h*k3}
    k4 = truncate(fx.subs(values),n)
    print(f"k4 = {temp.replace('t',f'({t}+{h})').replace('y',f'({y}+{h}*{k3})')}")
    printt(f"   = {temp.replace('t',f'({truncate(t+h,n)})').replace('y',f'({truncate(y+h*k3,n)})')} = {k4}",color='blue')
    ks.append(k4)

  return ks
```

```
def rkmethod2ndorder(fx,h,xvals,interval,n):
  a,b = interval
  N = round((b-a)/h)
  ys =  [xvals[1]]
  ts = [a]
  #dictt = {'y':f'yi', 't':f'ti'}
  for i in range(N):ts.append(truncate(ts[-1]+h,n))
  print(f'h = (b-a)/N =>   N = ({b}-{a})/{h} => {N}')
  [print(f't{i}={ts[i]}, ',end='') for i in range(len(ts))]
  print(f'\nf(t,y) = {fx}')
  printt(f'\n2nd Order RK method is\ny({xvals[0]}) = y(t0) = y0 = {ys[0]}\nyi+1 = y(ti+1) = yi + h*(f(ti+h/2, yi+(h/2)*k1))\n       where k1 = f(ti, yi)\n')

  temp = str(fx)  
  stringg = 'yi + h*(f(ti+h/2, yi+(h/2)*k1))'
  for i in range(1,N+1):
    printt(f'i={i})')
    ks = getks(fx,h,ts[i-1],ys[i-1],n,order=2)
    values = {'t':ts[i-1]+h/2,'y':ys[i-1]+(h/2)*ks[0]}
    ans = truncate(ys[i-1]+h*fx.subs(values),n)

    temp1 = stringg.replace('yi',str(ys[i-1])).replace('ti',str(ts[i-1])).replace('h',str(h)).replace('k1',str(ks[0]))
    temp2 = temp.replace('t',f'({truncate(ts[i-1]+h/2,n)})').replace('y',f'({truncate(ys[i-1]+(h/2)*ks[0],n)})')

    print(f"y{i} = {stringg.replace('i',str(i-1))}\n   = {temp1}")
    print(f"   = {ys[i-1]} + {h}*({temp2}) = {ans}")
    ys.append(ans)


  printt('\nSolution')
  print('t',end='\t')
  [print(i,end='\t') for i in ts]
  print('\ny',end='\t')
  [print(i,end='\t') for i in ys]
```

```
def rkmethod3rdorder(fx,h,xvals,interval,n):
  a,b = interval
  N = round((b-a)/h)
  ys =  [xvals[1]]
  ts = [a]
  #dictt = {'y':f'yi', 't':f'ti'}
  for i in range(N):ts.append(truncate(ts[-1]+h,n))
  print(f'h = (b-a)/N =>   N = ({b}-{a})/{h} => {N}')
  print(f'f(t,y) = {fx}')
  printt(f'\n3rd Order RK method is\ny({xvals[0]}) = y(t0) = y0 = {ys[0]}\nyi+1 = y(ti+1) = yi + (h/6)*(k1+4*k2+k3)\n',end='')
  printt('       where k1 = f(ti, yi)\n             k2 = f(ti+h/2, yi+(h/2)*k1)\n             k3 = f(ti+h, yi-h*k1+2*h*k2)\n')

  for i in range(1,N+1):
    printt(f'i={i})')
    ks = getks(fx,h,ts[i-1],ys[i-1],n,order=3)
    ans = truncate(ys[i-1] + (h/6)*(ks[0]+4*ks[1]+ks[2]),n)

    print(f"y{i} = y{i-1} + (h/6)*(k1+4*k2+k3)")
    print(f"   = {ys[i-1]} + ({h}/6)*({ks[0]}+4*{ks[1]}+{ks[2]})")
    print(f"   = {ys[i-1]} + {truncate(h/6,n)}*({truncate(ks[0]+4*ks[1]+ks[2],n)}) = {ans}")
    ys.append(ans)


  printt('\nSolution')
  print('t',end='\t')
  [print(i,end='\t') for i in ts]
  print('\ny',end='\t')
  [print(i,end='\t') for i in ys]
```

```
def rkmethod4thorder(fx,h,xvals,interval,n):
  a,b = interval
  N = round((b-a)/h)
  ys =  [xvals[1]]
  ts = [a]
  #dictt = {'y':f'yi', 't':f'ti'}
  for i in range(N):ts.append(truncate(ts[-1]+h,n))
  print(f'h = (b-a)/N =>   N = ({b}-{a})/{h} => {N}')
  print(f'f(t,y) = {fx}')
  printt(f'\n4th Order RK method is\ny({xvals[0]}) = y(t0) = y0 = {ys[0]}\nyi+1 = y(ti+1) = yi + (h/6)*(k1+2(k2+k3)+k4)\n',end='')
  printt('       where k1 = f(ti, yi)\n             k2 = f(ti+h/2, yi+(h/2)*k1)\n             k3 = f(ti+h/2, yi+(h/2)*k2)\n             k4 = f(ti+h, yi+h*k3)\n')

  for i in range(1,N+1):
    printt(f'i={i})')
    ks = getks(fx,h,ts[i-1],ys[i-1],n,order=4)
    ans = truncate(ys[i-1] + (h/6)*(ks[0]+2*ks[1]+2*ks[2]+ks[3]),n)

    print(f"y{i} = y{i-1} + (h/6)*(k1 + 2(k2+k3) + k4)")
    print(f"   = {ys[i-1]} + ({h}/6)*({ks[0]} + 2({ks[1]}+{ks[2]}) + {ks[3]})")
    print(f"   = {ys[i-1]} + {truncate(h/6,n)}*({truncate(ks[0]+2*ks[1]+2*ks[2]+ks[3],n)}) = {ans}")
    ys.append(ans)


  printt('\nSolution')
  print('t',end='\t')
  [print(i,end='\t') for i in ts]
  print('\ny',end='\t')
  [print(i,end='\t') for i in ys]
```
