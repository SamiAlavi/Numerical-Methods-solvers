```
from math import ceil, pi, e
from sympy import series, Symbol, Rational, diff, factorial, fraction, nsimplify, solveset, S, ln
from sympy.functions import sin, cos, tan
```

```
def truncate(floatt, n):
  '''truncate floatt to n digits'''
  return int(floatt * 10**n)/10.0**n

def getderivatives(fx, x):
  return abs(diff(fx, x, 1)) # first derivative

def choosefunc(fx1, fx2, a, b):
  dfx1 = getderivatives(fx1, x)
  dfx2 = getderivatives(fx2, x)
  
  if (dfx1.subs(x,a)<1 and dfx1.subs(x,b)<1) :
    return fx1
  elif (dfx2.subs(x,a)<1 and dfx2.subs(x,b)<1) :
    return fx2

def fixedpointitermethod(gx, initialguess, typee, stoppingcrit, limititer): 
  '''
  gx = equation
  initialguess
  typee (1,accuracy) (2,absolute error) (3, relative error)
  stoppingcrit for typee 1
  limititer = num of iters'''

  print('I\tx\tgx\tError')
  errorPn, errorPn1 = 0, 0

  for i in range(1,limititer+1):
    gxn = truncate(gx.subs(x, initialguess), n)

    errorPn = errorPn1
    errorPn1 = gxn

    if typee==1:
      error = (stoppingcrit)
    if typee==2:      
      error = (round(abs(errorPn1-errorPn), n), stoppingcrit)
    if typee==3:      
      error = (truncate(abs((errorPn1-errorPn)/errorPn1), n), stoppingcrit)
    print(f'{i}\t{initialguess}\t{gxn}\t{error}')

    initialguess = gxn
```

```
x = Symbol('x')

fx1 = 2**(-x)
fx2 = -ln(x,2)
a,b = 0,1
n = 3
initialguess = 0

stoppingcrit = 0.5*10**(2-n)
limititer = 8
typee = 3

gx = choosefunc(fx1, fx2, a, b)

fixedpointitermethod(gx, initialguess, typee, stoppingcrit, limititer)
```
