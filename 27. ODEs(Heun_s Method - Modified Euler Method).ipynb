```
from math import ceil, pi, e
from sympy import Symbol, diff, factorial, fraction, simplify, solve, N, symbols
from sympy.functions import sin, cos, tan
from termcolor import colored
```

```
def checkint(floatt):
  floatt = float(floatt)
  if floatt%1!=0:
    return floatt
  else:
    return int(floatt)

def truncate(floatt,n):
  floatt = str(float(floatt))
  indx = floatt.find('.')
  if indx!=-1:
    try:
      return checkint(floatt[:indx+n+1])
    except:
      return checkint(floatt)
  else:
    return checkint(floatt)

def printt(string,color="red",end='\n'):
  print(colored(string,color),end=end)
```

```
def modifiedeulermethod(fx,h,xvals,interval,n):
  a,b = interval
  N = round((b-a)/h)
  ys =  [xvals[1]]
  ts = [a]
  for i in range(N):ts.append(truncate(ts[-1]+h,n))
  print(f'h = (b-a)/N =>   N = ({b}-{a})/{h} => {N}')
  [print(f't{i}={ts[i]}, ',end='') for i in range(len(ts))]
  print(f'\nf(t,y) = {fx}\n')


  for i in range(1,N+1):
    printt(f'i = {i})')
    dictt = {'y':f'y{i-1}', 't':f't{i-1}'}
    dictvals = {'y':ys[i-1], 't':ts[i-1]}

    temp = fx.subs(dictt)
    val = truncate(fx.subs(dictvals),n)
    ys.append(truncate(ys[i-1] + h*val,n))

    # getting value of yi(1)
    print(f'y{i}(1) = y(t{i}) = y{i-1} + ({h})({temp}) ',end='')
    printt(f"=> {ys[i-1]} + ({h})({str(fx).replace('y',str(ys[i-1])).replace('t',str(ts[i-1]))}) ",color='blue',end='')
    print(f"=> {ys[i-1]} + ({h})({val}) ",end='')
    printt(f'= {ys[-1]}',color='green')

    # getting value of f(ti-1,yi-1) and f(ti,yi(1))
    dictt = {'y':f'y{i}(1)', 't':f't{i}'}
    dictvals = {'y':ys[-1], 't':ts[i]}
    fty1 = [fx.subs(dictt),truncate(fx.subs(dictvals),n)]
    print(f"f(t{i-1},y{i-1}) = {temp} = {str(fx).replace('y',str(ys[i-1])).replace('t',str(ts[i-1]))} = {val}")
    print(f"f(t{i},y{i}(1)) = {fty1[0]}  = {str(fx).replace('y',str(ys[-1])).replace('t',str(ts[i]))} = {fty1[1]}")

    # getting value of yi
    print(f'\ny{i} = y{i-1} + h((f(t{i-1},y{i-1}) + f(t{i},y{i}(1)))/2)')
    print(f'y{i} = {ys[i-1]} + ({h})(({val} + {fty1[1]})/2)')
    ys[i] = truncate(ys[i-1] + h*((val+fty1[1])/2),n)
    printt(f'y{i} = {ys[i]}\n','green')


  printt('Solution')
  print('t',end='\t')
  [print(i,end='\t') for i in ts]
  print('\ny',end='\t')
  [print(i,end='\t') for i in ys]
```

```
y,t = symbols('y t')    # set symbols

fx = (y-t)/(t+y)       # equation
h = 0.5
xvals = (0,1)         # y(0) = 1
interval = (0,0.5)        # set interval
n = 4                   # set decimals to truncate

modifiedeulermethod(fx,h,xvals,interval,n)
```
