```
from math import ceil, pi, e
from sympy import series, Symbol, Rational, diff, factorial, fraction, nsimplify, solveset, S, ln, expand, together, N
from sympy.functions import sin, cos, tan
from termcolor import colored
```

```
def printt(string, color, end='\n'):
  print(colored(string,color),end=end)

def fitlagrange(coords, x, xval, difftimes):
  Lx = list()
  strilist = list()
  for i in range(len(coords)):
    tempnum = 1
    tempden = 1
    string = ''
    for j in range(len(coords)):
      if j!=i:
        tempnum = tempnum * (x - coords[j][0])
        tempden = tempden * (coords[i][0] - coords[j][0])
        string += f'({coords[i][0]} - {coords[j][0]})'
    Lx.append(tempnum/tempden)
    strilist.append(f'{tempnum}/{string}')
  printt('Lx','red')
  [print(f"Lx{i} = {strilist[i]}\t{colored(f'=> {Lx[i]}','blue')}") for i in range(len(Lx))]

  printt('\nLx (expanded)','red')
  Lx = [expand(i) for i in Lx]

  for i in range(len(Lx)):
    temp = fraction(together(Lx[i]))
    print(f'Lx{i} = ({expand(temp[0])}) /{temp[1]}')

  dLxs = Lx
  for deriv in range(1,difftimes+1):
    dLxs = [nsimplify(diff(i,x,1)) for i in dLxs]
    apos = deriv*"'"

    printt(f'\n{deriv}) Derivatives Lx','red')
    if xval==None:
      for i in range(len(dLxs)):
        temp = fraction(together(dLxs[i]))
        print(f'L{apos}x{i}/d{x} = ({expand(temp[0])}) /{temp[1]}')
    else:
      [print(f"L{apos}x{i} = {dLxs[i]} {colored(f'=> {dLxs[i].subs(x,xval)}','blue')}") for i in range(len(Lx))]

  Px = list()
  for i in range(len(coords)):
    Px.append(dLxs[i] * coords[i][1])

  printt(f'\ndPx/d{x} (expanded)\ndPx{len(Px)-1}/d{x} = ','red',end='')

  if xval == None:
    [print(f'({dLxs[i]})({coords[i][1]}) +',end=' ') for i in range(len(dLxs))]
    print()
    for i in range(len(Px)):
      temp = fraction(together(Px[i]))
      print(f'Px{i} = ({temp[0]}) /{temp[1]}')
  else:
    [print(f'({dLxs[i].subs(x,xval)})({coords[i][1]}) +',end=' ') for i in range(len(dLxs))]
    print()
    [print(f'Px{i} = ',Px[i].subs(x,xval)) for i in range(len(Px))]

  ans = sum(Px)
  if xval == None:
    printt(f'\n\nPx = \t{N(ans)}','red')
    print(f'Px = \t{expand(ans)}')
  else:
    ans = ans.subs(x,xval)
    printt(f'\n\nPx = \t{ans} => {float(ans)}','red')
    
```

```
x = Symbol('t')

coords = [(0,0), (25,32), (50,58), (75,78)]
xval = None
difftimes = 1

fitlagrange(coords, x, xval, difftimes)
```
