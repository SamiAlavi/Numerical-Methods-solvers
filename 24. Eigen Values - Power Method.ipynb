```
from math import ceil, pi, e
from termcolor import colored
import numpy as np
```

```
def checkint(floatt):
  floatt = float(floatt)
  if floatt%1!=0:
    return floatt
  else:
    return int(floatt)

def truncate(floatt,n):
  floatt = str(float(floatt))
  indx = floatt.find('.')
  if indx!=-1:
    try:
      return checkint(floatt[:indx+n+1])
    except:
      return checkint(floatt)
  else:
    return checkint(floatt)

def printt(string,color="red",end='\n'):
  print(colored(string,color),end=end)
```

```
def finderror(lmbdas,k,n):
  a1, a2 = lmbdas[k-1], lmbdas[k]
  error = truncate(abs((a2-a1)/a2),n)
  print(f'Error = | ({a2} - {a1})/{a2} | => {error}\n')
  return error

def powermethod(A,x0,errorlim,n,numiters):
  flag = errorlim !=None
  rows = len(A)
  A, x0 = np.array(A), np.array(x0)
  xlist = [0,x0]
  lmbdas = [0]
  for k in range(1,numiters+1):
    printt(f'Ax{k-1}')
    dot = A.dot(xlist[k])
    dot = [truncate(i,n) for i in dot]
    lmbdas.append(max(dot, key=abs))
    dotdiv = np.divide(dot,lmbdas[k])
    dotdiv = [truncate(i,n) for i in dotdiv]
    for i in range(rows):
      if i==0:
        print(f'Ax{k-1} => {A[i]}[{xlist[k][i]}] = [{dot[i]}] = {lmbdas[k]}[{dotdiv[i]}]')
      else:
        print(f'       {A[i]}[{xlist[k][i]}]     [{dot[i]}]     [{dotdiv[i]}]')
    xlist.append(dotdiv)
    printt(f'x{k} = {dotdiv}t, λ{k} = {lmbdas[k]}',color='blue')
    error = finderror(lmbdas,k,n)
    if flag:
      if error<errorlim:
        break
  
  printt(f'{lmbdas[-1]} is the largest eigen value with eigen vector {xlist[-1]}t with error {error}','green')

```

```
A = [[1,2],         # matrix A
     [3,2]]

x0 = [1,1]            # initial x0
errorlim = 5/100       # if errorlim, else set to None
n = 4                 # for truncation (n decimal places)
numiters = 50          # num of iterations (set to 50 if to find using errorlim)

powermethod(A,x0,errorlim,n,numiters)
```

```
A = [[1,1,0,0],         # matrix A
     [1,2,0,1],
     [0,0,3,3],
     [0,1,3,2]]

x0 = [1,1,0,1]          # initial x0
errorlim = None         # if errorlim, else set to None
n = 4                  # for truncation (n decimal places)
numiters = 3            # num of iterations (set to 50 if to find using errorlim)

powermethod(A,x0,errorlim,n,numiters)
```
