```
from math import ceil, pi, e
from sympy import Symbol, symbols, solve, Matrix, pretty, pprint, expand, N, factor, re
from sympy.solvers.solveset import linsolve
from termcolor import colored
import numpy as np
```

```
def truncate(floatt,n=4):
  floatt = str(float(floatt))
  indx = floatt.find('.')
  if indx!=-1:
    try:
      return float(floatt[:indx+n+1])
    except:
      return float(floatt)
  else:
    return float(floatt)

def printt(string,color="red",end='\n'):
  print(colored(string,color),end=end)

def geturl(st):
  temp1 = str(st).replace('**','^').replace(' ','').replace('*','')    
  print(f'\n(For steps) Paste this ({temp1}) in \nhttps://www.symbolab.com/solver/simplify-calculator/simplify')

def directmethod(A):
  printt('AX = λX\nAX - λX = 0\n(A - λI)X = 0\nTo find eigen values,solve\n|A-λI| = 0')
  rows = len(A)
  A = np.array(A)
  lmbda = Symbol('λ')
  I = np.identity(rows,dtype = int)
  temp1 = Matrix(A)
  temp2 = Matrix(I)
  X = list()
  for i in range(rows):
    X.append(Symbol(f'x{i+1}'))
  symbs = X
  X = Matrix(X)

  for i in range(rows):
    if i==0:
      print(f'\n=> | {pretty(temp1.row(i))}   - λ{pretty(temp2.row(i))} | = 0')
    else:
      print(f'   | {pretty(temp1.row(i))}     {pretty(temp2.row(i))} |')

  I = lmbda*I
  temp2 = Matrix(I)

  for i in range(rows):
    if i==0:
      print(f'\n=> | {pretty(temp1.row(i))}   - {pretty(temp2.row(i))} | = 0')
    else:
      print(f'   | {pretty(temp1.row(i))}    {pretty(temp2.row(i))} |')
  
  A = A - I
  A = Matrix(A)
  for i in range(rows):
    if i==0:
      print(f'\n=> | {pretty(A.row(i))}  | = 0')
    else:
      print(f'   | {pretty(A.row(i))} |')
  
  det = A.det()
  values = solve(det)
  print(f'\n{det} = 0')
  det = expand(det)  
  print(f'{det} = 0 (Characteristic polynomial)')
  values = [truncate(re(N(val))) for val in values]
  print(f'{factor(det)} = 0')
  print(f'λ = {values}')
  geturl(det)

  for val in values:
    printt(f'\n----------------------------------\nλ = {val} (A - λI)X = 0')
    temp1 = A.subs(lmbda,val)
    for i in range(rows):
      if i==0:
        print(f'=> {pretty(temp1.row(i))}  {pretty(X.row(i))} = [0]')
      else:
        print(f'   {pretty(temp1.row(i))}  {pretty(X.row(i))}   [0]')

    dotprod = temp1.dot(X)
    varvals = [i for i in list(linsolve(dotprod,symbs))[0]]
    varvals = dict(zip(symbs,varvals))

    for i in range(rows):
      if i==0:
        print(f'\n=> [{dotprod[i]}] = [0]')
      else:
        print(f'   [{dotprod[i]}]   [0]')

    print(f"\n{dotprod[0]} = 0 {colored(f'=> {symbs[1]} = {varvals}','blue')}")
    temp1 = X.subs(varvals)
    temp2 = temp1/symbs[-1]
    flag = True
    while flag:
      for i in range(rows):
        if temp2[i]%1!=0:
          temp2*=2
          flag = True
          break
        else:
          temp2[i] = int(temp2[i])
        flag = False
        
    for i in range(rows):
      if i==0:
        print(f'[{X[i]}] = [{temp1[i]}] = [{temp2[i]}]')
      else:
        print(f'[{X[i]}]   [{temp1[i]}]   [{temp2[i]}]')

    printt(f'λ = {val}, eigen vector =')
    pprint(temp2)
```

```
A = [[1,1],
     [-2,4]]

directmethod(A)
```

```
A = [[8,-8,-2],
     [4,-3,-2],
     [3,-4,1]]

directmethod(A)
```
