```
def truncate(floatt):
  '''truncate floatt to n digits'''
  return round(round(floatt,10)*100000)/100000.0

def findmid(val):
  n = len(val) 
  mid = n//2    
  if n % 2 == 0: 
      median = (val[mid] + val[mid - 1])/2
  else: 
      median = val[mid]
  return median

def stirlingcentdiff(coords, xs, x):
  values = [[i[1] for i in coords]]
  length = len(coords)
  h = truncate(xs[1] - xs[0])
  flag = x==None
  if flag:
    x = Symbol('x')
    s = (x - findmid(xs))/h
  else:
    s = truncate((x - findmid(xs))/h)
  print(f's = (x-x0)/(h)')
  print(f's = ({x}-{findmid(xs)})/({xs[1]}-{xs[0]})\ns = {s}')

  for i in range(len(coords)-1):
    temp = list()
    orders = values[-1]
    for j in range(len(orders)-1):
      temp.append(truncate(orders[j+1]-orders[j]))
    values.append(temp)

  # creating table
  data = {'x':xs,'y':values[0]}
  [data.update({i:values[i]}) for i in range(1,length)]

  df = pd.DataFrame.from_dict(data, orient='index')
  df = df.fillna('-').transpose()
  display(HTML(df.to_html(index=False))) # table created

  if flag:  # for equation finding
    findequation(values, s, length)
  else:  # for solution finding
    findsolution(values, s, length, x)

def findsolution(values, s, length, x):
  ylist = [findmid(values[0])]
  slist = [1,s]
  temp = [1,'(s)']

  for i in range(1,length):
    ylist.append(findmid(values[i]))
    if i % 2 == 0:
      slist.append(slist[-2]*(s**2 - (i//2)**2))
      temp.append(temp[-2]+f'(s**2 - {i//2}**2)')
    else:
      slist.append(slist[-1]*s)
      temp.append('(s)'+temp[-1])
  temp, slist = temp[:-1], slist[:-1]

  print('f(x) =\t(y0) + \n\t',end='')
  for i in range(1,length):
    print(f'({temp[i]}/{i}!)(δ**{i}y) + \n\t',end='')
  
  print(f'\nf({x})=({ylist[0]}) + \n\t',end='')
  for i in range(1,length):
    print(f'({slist[i]}/{i}!)({ylist[i]}) + \n\t',end='')

  ad = [ylist[0]]
  #print(f'\nf({x})=({ad[-1]}) + \n\t',end='')
  for i in range(1,length):
    ad.append(slist[i]*ylist[i]/factorial(i))
    #print(f'({ad[-1]}) + \n\t',end='')
  print(f'\nf({x}) = {sum(ad)}')
```
