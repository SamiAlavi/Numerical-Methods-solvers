```
from math import sqrt, e
from sympy import series, Symbol, Rational, diff, factorial, fraction, nsimplify
from sympy.functions import sin, cos, tan
```

```
def taylor(fx, x0, n):
    """
    Parameter "function" is our function which we want to approximate
    "x0" is the point where to approximate
    "n" is the order of approximation
    """
    return nsimplify(fx.series(x,x0,n+1).removeO())

def getderivatives(fx, x, n):
  a = fx
  for i in range(1,n+2):
    a = diff(a, x, 1) #derivative
    print(f'{i} derivative = {a}')    
  print()
  return a

def getR(n, xval, der):
  numerator, denominator = fraction(der.subs(x,0))
  print(f'{xval**(n+1)*numerator}/{denominator*factorial(n+1)}')
```

Q) Calculate Maclaurin’s series for 𝒆^𝒙. Calculate 𝒆 with an error of less than 𝟏𝟎^(−𝟔).

```
def findN(fx, xval, maxnval, errorval):
  numerator = round(fx.subs(x,xval))
  for i in range(maxnval):
    denominator = factorial(i+1)
    val = float(numerator/denominator)
    flag = val<errorval
    print(f'n = {i}\t{val}\t{errorval}\t{flag}')
    if flag:
      print()
      return i
```
